```python
import random

def sensor_presenca():
    return random.choice([True, False])

def controle_iluminacao(presenca):
    return "Ligada" if presenca else "Desligada"

print("=== Simulação de IoT - Iluminação Inteligente ===")
for i in range(5):
    presenca = sensor_presenca()
    estado = controle_iluminacao(presenca)
    print(f"Sensor {i+1}: presença = {presenca} -> luz {estado}")
```

## Análise de Dados de Consumo de Energia em Escritórios Híbridos: A Relevância da Simulação IoT

Esta simulação demonstra um conceito fundamental para a gestão de energia em ambientes de escritórios híbridos: a **automação baseada na presença**.

### O que a Simulação Faz?

O código atual simula um sistema de iluminação inteligente simples:

*   **`sensor_presenca()`**: Representa um sensor que detecta se há alguém em uma área (retornando `True` para presença ou `False` para ausência).
*   **`controle_iluminacao(presenca)`**: Decide se a luz deve ser ligada (`Ligada`) ou desligada (`Desligada`) com base na informação do sensor de presença.
*   A simulação executa esse ciclo várias vezes, mostrando como a luz se adapta dinamicamente à 'presença' no ambiente.

### Conexão com Escritórios Híbridos e Consumo de Energia


1.  **Detecção de Ocupação**: Em escritórios híbridos, a ocupação dos espaços varia muito. Sensores de presença (como o simulado aqui) ou outros dados (agendamentos de salas, logins em computadores) são essenciais para entender quando e onde as pessoas estão presentes.
2.  **Otimização do Consumo**: A iluminação é um grande consumidor de energia. Ao integrar dados de ocupação (como os `presenca` da simulação) com sistemas de controle, é possível otimizar o uso da iluminação e climatização, garantindo que a energia só seja consumida quando e onde necessário.
3.  **Base para Análise**: Os dados gerados por esses sistemas (presença vs. estado da luz/energia) são os "dados de consumo" que vocês analisarão. Por exemplo, podem comparar:
    *   Consumo de energia em salas com e sem sensores de presença.
    *   Padrões de ocupação em diferentes horários e dias da semana.
    *   A economia potencial ao automatizar sistemas com base na ocupação.

Esta simulação serve como um modelo conceitual básico para entender como a coleta de dados de um "sensor" pode levar a decisões automatizadas que impactam diretamente o consumo de energia em um cenário de escritório híbrido.
